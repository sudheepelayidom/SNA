```python

#Prestige computation

def directed_graph():
  # to construct a graph using lists
  G2 = nx.DiGraph()

  # creating nodes and edges

  nodes = [1,2,3,4]
  n_nodes = len(nodes)
  edges = [(1,2),(1,3),(2,1),(2,3),(3,1),(3,4)]

  G2.add_edges_from(edges)

  # to display the network
  nx.draw_networkx(G2,node_color='orange',node_size=2000)

  return G2,nodes, n_nodes, edges
  # adjancency matrix
Adj = nx.adjacency_matrix(G2)
print(Adj.todense())
G2,nodes,n_nodes,edges = directed_graph()
degree_prestige = dict((v,len(G2.in_edges(v))/(n_nodes-1)) for v in G2.nodes)

print("\n Degree Prestige: \n")
for cnt_Dp in degree_prestige:
  print(cnt_Dp," : ", degree_prestige[cnt_Dp])
```

```python
#proximity prestige

G2 = nx.DiGraph()
# creating nodes and edges

nodes = [1,2,3,4]
n_nodes = len(nodes)
edges = [(1,2),(1,3),(2,1),(2,3),(3,1),(3,4)]
G2,nodes,n_nodes,edges = directed_graph()
# for proximity prestige
distance = []; temp_dis = 0; n = 0
for dest in G2.nodes:
    temp_dis = 0; n = 0
    for src in G2.nodes:
        if (nx.has_path(G2,src,dest) == True):
            temp_dis = temp_dis + nx.shortest_path_length(G2,source = src,target = dest)
            n = n + 1
        if temp_dis == 0:
            distance.append([dest, 0])
        else:
            distance.append([dest, temp_dis/(n - 1)])
            print("\n Proximity Prestige :\n")
        for i in distance:
            print(str(i[0]) + " : " + str(i[1]))
```

```python
# to calculate distance in a network
# to display the network
nx.draw_networkx(G1,node_color='orange',node_size=2000)

# to measure eccentricity (largest d/s b/w nodes i to all other nodes in the graph )
paths = nx.eccentricity(G1)
print(paths)

# to measure the diameter (maximum eccentricity)
paths_rad = nx.diameter(G1)
print(paths_rad)

# to measure radius (minimum eccentricity)
paths_dia = nx.radius(G1)
print(paths_dia)
```
