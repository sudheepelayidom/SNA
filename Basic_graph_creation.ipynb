```python
import networkx as nx
```

```python
G=nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_node(2)
```

```python
G.add_node(3)
```

```python
G.add_node(4)
```

```python
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)
G.add_edge(3,4)
```

```python
nx.draw_networkx(G,node_color='orange',node_size=2000)
```
